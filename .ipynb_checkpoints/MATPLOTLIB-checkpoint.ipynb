```python
import matplotlib.pyplot as plt
import pandas as pd
```

## Bar Chart

```python
x=['python','c','c++','java']
y=[80,65,75,93]
plt.xlabel('Language',fontsize=15)
plt.ylabel('Percentage',fontsize=15)
plt.title('Bar Graph',fontsize=15)
c=["y","b","m","g"]
plt.bar(x,y,width=0.4,color=c)
plt.show()
```

## 2D plot

```python
price=[48000,54000,57000,47000,45000]
year=[2015,2016,2017,2018,2020]
plt.plot(year,price)
```

```python
data=pd.read_csv('matches - matches.csv')
```

```python
plt.plot(data['season'],data['team1'])
plt.plot(data['season'],data['team2'])
plt.title('Matches against team1 & team2')
plt.xlabel('Year',color="b",fontsize=15)
plt.ylabel('Teams Name',fontsize=15,color="g")
```
