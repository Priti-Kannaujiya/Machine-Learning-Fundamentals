```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
plt.style.use('fivethirtyeight')
```

## line plot

```python
data=sns.load_dataset("penguins")
```

```python
data.head()
```

```python
sns.lineplot(x='bill_length_mm',y='bill_depth_mm',data=data,hue='sex',style='sex',markers=['o','>'],dashes=False)
```

## Scatter plot

```python
data=sns.load_dataset('tips')
data
```

```python
sns.relplot(x='total_bill',y='tip',data=data)            # kind='scatter' bydefault hota hai but for other graphs we need to change
```

```python
# hue PARAMETER
sns.relplot(x='total_bill',y='tip',data=data,hue='smoker',style='sex',size='size',sizes=(15,100)) 
```

```python
sns.scatterplot(x='total_bill',y='tip',data=data,hue='smoker',style='sex',size='size',sizes=(15,100)) # we can show the above graph using scatter plot
                                                                                                      # function also
```

```python
sns.catplot(x='day',y='tip',kind='strip',jitter=0,data=data)
#jitter
```

```python
sns.catplot(x='day',y='tip',kind='swarm',data=data,hue='sex')
plt.show()
```

```python
sns.swarmplot(x='day',y='tip',data=data,hue='sex')
```

## Histogram

```python
var=sns.load_dataset("tips")
```

```python
sns.displot(var['total_bill'],bins=[10,20,30,40,50,60],kde=True,rug=True)
plt.show()
```

## BAR PLOT

## BOX PLOT

```python
var=sns.load_dataset("tips")
var
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",hue="sex",data=var,showmeans=True,
           meanprops={"marker":"*","markeredgecolor":"g"},linewidth=2,palette='plasma',orient="v")                                                                                              #hue="sex"
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(y=var["total_bill"])
```

## VIOLIN PLOT

```python
var=sns.load_dataset("tips")
```

```python
var
```

```python
sns.violinplot(x="smoker",y="total_bill",data=var,hue="sex",palette="crest",order=["No","Yes"],saturation=0.5,split=True)
plt.show()
```

```python
sns.violinplot(x="smoker",y="total_bill",data=var,inner="quart",color="r")
plt.show()
```

```python
sns.violinplot(x=var["total_bill"])
```

## PAIR PLOT

```python
sns.pairplot(var,hue="sex",hue_order=["Female","Male"],palette="BuGn",kind="scatter",markers=["*","<"])
plt.show()
```
