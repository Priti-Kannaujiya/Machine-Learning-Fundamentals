```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Bar Chart

```python
x=['python','c','c++','java']
y=[80,65,75,93]
plt.xlabel('Language',fontsize=15)
plt.ylabel('Percentage',fontsize=15)
plt.title('Bar Graph',fontsize=15)
c=["y","b","m","g"]
plt.bar(x,y,width=0.4,color=c)
plt.show()
```

```python
df=pd.read_csv('batsman_season_record.csv')
plt.barh(df['batsman'],df['2015'],color='#33ffce')
```

```python
# STACK BAR CHART
plt.bar(df['batsman'],df['2015'],label='2015')
plt.bar(df['batsman'],df['2016'],bottom=df['2015'],label='2016')
plt.bar(df['batsman'],df['2017'],bottom=(df['2015']+df['2016']),label='2017')
plt.legend()
```

```python
plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width=0.2,label='2015')           # way to represent multiple bar graph
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,label='2016')
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width=0.2,label='2017')
plt.legend()
plt.xticks(np.arange(df.shape[0]),df['batsman'])
plt.show()
```

## 2D plot

```python
price=[48000,54000,57000,47000,45000]
year=[2015,2016,2017,2018,2020]
plt.plot(year,price)
```

```python
data=pd.read_csv('sharma-kohli.csv')
```

```python
plt.plot(data['index'],data['RG Sharma'],linestyle='dashed')
plt.plot(data['index'],data['V Kohli'],linestyle='dashdot',linewidth=2)
plt.title('V Kohli v/s RG Sharma')
plt.xlabel('Year',color="b",fontsize=15)
plt.ylabel('Runs',fontsize=15,color="g")
```

```python
plt.plot(data['index'],data['RG Sharma'],linestyle='dashed',marker='D',label='RG Sharma')
plt.plot(data['index'],data['V Kohli'],linestyle='dashdot',linewidth=2,marker='*',label='V Kohli')
plt.title('V Kohli v/s RG Sharma')
plt.xlabel('Year',color="b",fontsize=15)
plt.ylabel('Runs',fontsize=15,color="g")
plt.legend()                                     # it is used to mark the graph:which graph represent which team
plt.grid()                                       # this help in readability of graph
```

```python
price=[48000,54000,57000,47000,4500000,45000]
year=[2015,2016,2017,2018,2020,2021]
plt.plot(year,price)
plt.ylim(40000,75000)            # this is used to trim the graph when an outlier value exist    
plt.show()             # it is used to  brings up the plot in a new window.Without plt.show(), plots might not display correctly, especially in scripts.
```

## Scatter Plot

```python
df=pd.read_csv('batter.csv')
df.head(10)
```

```python
plt.scatter(df['avg'],df['strike_rate'],color='g',marker='*')
plt.xlabel('Average',color='r',fontsize=15)
plt.ylabel('Strike Rate',color='r',fontsize=15)
plt.title('batsman strike rate v/s average rate',color='r',fontsize=15)
plt.grid()
```

```python
# SCATTER PLOT USING PLT.PLOT
# Faster than the scatter plot. Generally used in larger dataset
plt.plot(df['avg'],df['strike_rate'],'o',color='r')
```

## PIE CHART
